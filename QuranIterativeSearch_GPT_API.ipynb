# Quran Iterative Search Using OpenAPI GPT API

```python
import openai
import os, time, re
import pandas as pd

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

openai.api_key  = os.getenv('OPENAI_API_KEY')
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
```

```python
df_gaala = pd.DataFrame(columns=['Response', 'Occurrences'])
len(df_gaala.index)
```

```python
def get_results(df, prompt, iterations=1, pause=0):
    
    for i in range(iterations):
        # API call
        response = get_completion(prompt)

        # Extract the number of occurrences from response
        try:
            occurs = int(re.findall(r'\d+', response)[-1])
        #         occurs = int(response.split(sep='occurrences: ')[1])
        except ValueError:
            occurs = None
        except IndexError:
            print(f'Didn\'t find occurrences.\nThe response:\n{response}')

        # Append row
        df.loc[len(df.index)] = [response, occurs]

        # Wait before next API call
        time.sleep(pause)
        
    return df
```
